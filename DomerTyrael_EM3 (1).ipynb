```python
# write code here
# Make dictionaries for month entries in number form and incomplete months
months1 = {"1":"January","2":"February","3":"March","4":"April","5":"May","6":"June","7":"July","8":"August","9":"September","10":"October","11":"November","12":"December"}
months2 = {"Jan":"January","Feb":"February","Mar":"March","Apr":"April","May":"May","Jun":"June","Jul":"July","Aug":"August","Sep":"September","Oct":"October","Nov":"November","Dec":"December"}

# Define function to return complete month
def clean_month(month):
    if month in months1.keys():
        return months1[month]
    elif month in months2.keys():
        return months2[month]
    else:
        return month

#test on copy
copy = recweek.copy()
copy["birthmonth"] = copy["birthmonth"].apply(clean_month)
copy["birthmonth"]
```

```python
# Import datetime module (researched)
import datetime 

# Define function that considers both cases of YYYY and YY values, returns the year value in string type
def clean_year(year):
    year = str(year)
    if len(year) == 4 :
        dt = datetime.datetime.strptime(year, '%Y' )
        return dt.year
    elif len(year) == 2:
        dt_year = datetime.datetime.strptime(year,'%y')
        return dt_year.year
    

recweek['birthyear'] = recweek['birthyear'].apply(clean_year)

#Source: https://www.w3schools.com/python/python_datetime.asp
```

```python
# Create dictionary with course codes as keys and full course names as values
courses = {'CTM':['BS CTM','CTM','Communications Technology','BS Communications Technology'],
           'ITE':['ITE','BS ITE','Information Technology Entepreneurship','BS Information Technology Entrepreneurship'],
          'LM':['LM','BS LM','Legal Management','BS Legal Management'],
          'MGT':['MGT','BS MGT','Management','BS Management'],
          'ME':['ME','BS ME','Management Engineering','BS Management Engineering'],
          }

# Define function that cleans the values in course column based on dictionary above
def clean_course(course):
    if course in courses['CTM']:
        return 'BS Communications Technology'
    elif course in courses['ITE']:
        return 'BS IT Entrepreneurship'
    elif course in courses['LM']:
        return 'BS Legal Management'
    elif course in courses['MGT']:
        return 'BS Management'
    elif course in courses['ME']:
        return 'BS Management Engineering'
    
recweek['course'] = recweek['course'].apply(clean_course)
```

```python
# write code here

# Define function that cleans the phone numbers 
def clean_number(number):
    if number[0:4] == "+63 ":
        number = number.replace("+63 ","0")
        return number
    elif number[0:1] == "9":
        number = "0" + number[0:3] + "" + number[3:7] + "" + number[7:12] 
        return number
    else:
        return number

# Apply clean_number function to phonenumber column in recweek
recweek['phonenumber'] = recweek['phonenumber'].apply(clean_number)
recweek
```

```python
# write code here

copy_network1 = final_recweek.copy()

# Dictionary of network prefixes
networks={'Globe':['0917','0925','0905','0906','0915','0916','0926','0927','0935','0936','0937','0945','0953','0954','0955','0956','0965','0966','0967','0975','0977','0978','0979','0995','0996','0997'], 
        'Smart':['0908','0918','0919','0920','0921','0928','0929','0939','0946','0947','0949','0951','0961','0998','0999']}

# Function to get prefix of each number
def get_prefix(num):
    prefix = num[0:4]
    return prefix

# Function to get network of each prefix
def get_network(number):
  
    prefix = get_prefix(number)
    
    if prefix in networks['Globe']:
        return "Globe"
    elif prefix in networks['Smart']:
        return "Smart"

# Insert a new column of carriers to be able to use count
copy_network1.insert(2, 'Carrier', copy_network1['phonenumber'].apply(get_network))
copy_network2 = copy_network1.copy()

# Drop and rename columns
copy_network1 = copy_network1.drop(['lastname','firstname','birthday','birthmonth','yearlevel','course','birthyear','phonenumber'],axis=1)

count_carrier = copy_network1.groupby('Carrier').count()
count_carrier.rename(columns={'gender':'count'}, inplace=True)
```
